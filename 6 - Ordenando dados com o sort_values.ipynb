```python
#Importando pandas e abrindo o arquivo
import pandas as pd
df= pd.read_csv("houses_to_rent_v2.csv")
```

```python
# Usando o sort_values
df.sort_values("city").head()
```

```python
# Ordenando por mais de uma coluna
df.sort_values(["city", "rooms"]).head()
```

```python
# Ordenando com asc ou desc
df.sort_values("city", ascending=True).head()
```

```python
# Ordenando mais de uma coluna com ascending, passando parâmetros por dentro dos vetores.
df.sort_values(["city", "rooms"], ascending=[True, False]).head()
```

```python
# Ordenando colunas com passagem de parâmetro
colunas = ["city", "rooms"]
ordenacao = ["True", "False"]

df.sort_values(colunas, ascending= ordenacao).head()
```
